```python
from nltk import pos_tag
from nltk.corpus import stopwords
import string
stop = set(stopwords.words("english"))
punctuations = list(string.punctuation)
stop.update(punctuations)
```

```python
from nltk.corpus import wordnet
def get_simple_pos(pos_tg):
    if pos_tg.startswith('J'):
        return wordnet.ADJ
    if pos_tg.startswith('V'):
        return wordnet.VERB
    if pos_tg.startswith('N'):
        return wordnet.NOUN
    if pos_tg.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
        
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def clean(words):
    output_word = []
    for word in words:
        if word.lower() not in stop:
            pos_tg = pos_tag([word])
            clean_word = lemmatizer.lemmatize(word,pos=get_simple_pos(pos_tg[0][0]))
            output_word.append(clean_word)
    return output_word
            
        
```

```python
documents = [(clean(word),cat) for word,cat  in documents ]
```

```python
print(documents[10])
```

```python
training_data = documents[0:1500]
testing_data = documents[1500:]
```

```python
print(training_data[0])
```

```python
all_words=[]
for i in training_data:
    all_words+=i[0]
```

```python
freq = nltk.FreqDist(all_words)
```

```python
most_common=freq.most_common(3000)
```

```python
features = [i[0] for i in most_common]
```

```python
print(features)
```

```python
all_words[0:10]
```

```python
features
```

```python
def get_features_dict(words):
    current_features = {}
    word_set = set(words)
    for w in features:
        current_features[w] = w in word_set
    return current_features
```

```python
training_data = [(get_features_dict(words),cat) for words,cat in training_data]
```

```python
testing_data = [(get_features_dict(words),cat) for words,cat in testing_data]
```

```python
print(training_data[0])
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier =NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```
